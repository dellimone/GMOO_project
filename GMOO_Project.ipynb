# GMOO Project: Hybridization of PSO with GA

## Introduction

### Aim of the study

The aim of this study is to enhance Particle Swarm Optimization (PSO) by integrating key operations from Genetic Algorithms (GA). PSO excels at exploitation, allowing particles to converge quickly toward promising solutions, while GA is more effective at exploration, maintaining diversity in the solution space. By incorporating GA mechanisms such as selection, crossover, or mutation into PSO, the goal is to develop a more robust and efficient optimization algorithm that balances exploration and exploitation.

This hybrid approach is expected to improve convergence speed, prevent premature stagnation, and enhance solution quality in complex optimization problems.

### Study Structure

The project is structured into three main parts.

The first part focuses on implementations, introducing various hybridization techniques that combine Genetic Algorithms (GA) and Particle Swarm Optimization (PSO). Four different approaches are explored: Sequential GAPSO, which follows a step-by-step integration of GA and PSO; Alternating GAPSO, where the two algorithms operate in an alternating manner; Parallel GAPSO, where both algorithms run simultaneously; and Island GAPSO, which uses an island model where separate populations evolve independently with occasional migration of solutions.  

The second part is dedicated to comparisons, where the different hybrid approaches are tested using benchmark functions. This evaluation focuses on several key aspects, including the performance of the algorithms on standardized benchmark functions, the convergence history to assess how quickly and effectively each method reaches optimal solutions, and a detailed analysis of the quality and reliability of the obtained solutions.  

The final part consists of the conclusions, summarizing the key findings from the study. It also includes a discussion on future developments, highlighting potential improvements and directions for further research in hybrid GAPSO techniques. This structured approach ensures a comprehensive exploration of hybridization methods, from implementation to performance evaluation and future perspectives.

## Hybridization techniques

### GA

The process starts with generating an initial population of random solutions within defined bounds, and then refines these solutions over multiple generations.

For crossover, the algorithm uses **BLX-alpha**, a blend crossover technique where two parent solutions combine their features based on a random factor. This factor determines how much of each parent's characteristics influence the offspring. The offspring are created by blending the parent solutions, ensuring that the new solutions are not too far from the originals. The offspring are then clipped to stay within the problem's defined bounds.

For mutation, **Gaussian noise** is added to the solutions. In each generation, the algorithm applies a mutation to some individuals, adding small random changes based on a normal distribution. The mutation scale adapts over time, starting larger and gradually decreasing as the algorithm progresses, to balance exploration (searching new areas) with exploitation (refining existing good solutions). The mutated solutions are also clipped to ensure they stay within the defined bounds of the problem.

Elitism ensures that the best solutions are preserved and passed on to the next generation, maintaining high-quality solutions. The optimization process runs for a specified number of iterations, and the final result includes the best solution found, its fitness, and the convergence history.

### PSO

In Particle Swarm Optimization (PSO), a population of particles is initialized with random positions and velocities within the problem's search space. Each particle evaluates its fitness and tracks both its personal best and the global best. Over multiple iterations, the particles update their velocities based on their personal best and the global best, with a dynamic inertia weight that decreases over time. The new position of each particle is then calculated. The algorithm continues iterating until it reaches the maximum number of iterations, returning the best solution found by the swarm.

### Sequental GAPSO

Inspired by the paper [Hybrid PSO and GA for Global Maximization,K. Premalatha and A.M. Natarajan](https://www.emis.de/journals/IJOPCM/Vol/09/IJOPCM(vol.2.4.12.D.9).pdf)

 The **Sequential GAPSO (SGAPSO)** algorithm is a hybrid optimization approach that combines Genetic Algorithms (GA) and Particle Swarm Optimization (PSO) in a sequential manner. The process is divided into two distinct phases:  

First, the **GA phase** is executed, where the population undergoes genetic operations such as selection, crossover, and mutation to explore the solution space. GA helps maintain diversity in potential solutions, preventing premature convergence to local optima.  

Once the GA phase is complete, the best individuals from GA are passed to the **PSO phase**, where particles refine the solutions further by iteratively adjusting their positions based on personal and global best values. This phase enhances convergence efficiency, leveraging PSO’s strong exploitation capabilities to fine-tune the solutions obtained from GA.  

By structuring the optimization in two phases, SGAPSO takes advantage of GA’s exploration ability to generate diverse candidate solutions while benefiting from PSO’s fast convergence to refine and improve the best solutions found.

### Alternating GAPSO

Inspired by the paper [A hybrid PSO-GA algorithm for constrained optimization problems, Harish Garg](https://doi.org/10.1016/j.amc.2015.11.001)

The **Alternating GAPSO (AGAPSO)** algorithm is a hybrid optimization approach that interleaves steps of Particle Swarm Optimization (PSO) and Genetic Algorithm (GA) within each iteration. Instead of running these methods sequentially, APSOGA applies a PSO step followed by a GA step to a subset of the population, aiming to balance exploration and exploitation dynamically throughout the optimization process.  

In each iteration, **PSO updates** adjust the positions of particles based on their personal and global best solutions. This allows individuals to move towards promising regions of the search space. Following this, **GA operations** introduce additional diversity and refinement through selection, crossover, and mutation.  

For crossover, AGAPSO employs BLX-alpha crossover, which generates offspring by extrapolating beyond the parent solutions within a defined range. This mechanism encourages exploration by allowing offspring to inherit traits from both parents while potentially extending beyond their values, promoting genetic diversity.

**Mutation** applies Gaussian perturbations to individuals, adjusting their positions based on a dynamically decreasing mutation scale. This ensures more aggressive exploration in early iterations and finer adjustments as optimization progresses. 

By continuously alternating between PSO and GA within each iteration, AGAPSO maintains a strong balance between convergence speed and diversity, reducing the risk of premature convergence while efficiently navigating complex search spaces.

### Parallel GAPSO

Inspired by the paper [A hybrid genetic algorithm and particle swarm
optimization for multimodal functions,Yi-Tung Kao Erwie Zahara](http://dx.doi.org/10.1016/j.asoc.2007.07.002)

The Parallel GAPSO (**PGAPSO**) is a hybrid metaheuristic that combines Genetic Algorithm (GA) and Particle Swarm Optimization (PSO) in a parallelized manner to balance exploration and exploitation in optimization problems. The algorithm maintains a population of candidate solutions and splits it into two groups based on fitness rankings.  

- The **top-performing half** undergoes **Genetic Algorithm (GA) operations**, which include crossover and mutation, to promote diversity and enable global exploration.  
- The **bottom-performing half** is updated using **Particle Swarm Optimization (PSO) principles**, where solutions adjust their positions based on personal and global bests to enhance local exploitation.

### Island GAPSO

 This algorithm combines **Particle Swarm Optimization (PSO)** with **Genetic Algorithms (GA)** in an **Island Model** approach. Multiple independent PSO-based "islands" run in parallel, each optimizing a portion of the search space. Periodically, the best solutions found by the islands are exchanged using GA-based migration, which involves **tournament selection**, **simulated binary crossover (SBX)**, and **Gaussian mutation**. This migration introduces diversity by sharing successful solutions between islands, helping to explore new regions of the search space and avoid premature convergence.

Each island runs its own PSO to optimize the problem, and the best solution found across all islands is updated throughout the process. The GA migration step ensures that islands benefit from the solutions of others, enhancing both exploration (through diversity) and exploitation (through the use of the best solutions). The algorithm tracks the global best solution, as well as the best solution from each island, providing insights into the optimization process. Overall, the hybrid PSO-GA approach aims to improve optimization performance by combining the strengths of both algorithms.

### Ackley function

The **Ackley function** is a widely used benchmark function in optimization, known for its complex, multi-modal landscape. It has a global minimum at the origin $(0, 0, \dots 0)$ but features many local minima in its domain, making it challenging for optimization algorithms. 

The function is composed of two terms: one that creates a smooth, bowl-shaped surface and another that introduces oscillations. These combined characteristics give it a rugged and difficult-to-optimize surface. 

The general form of the Ackley function is:

$$
f(x) = -20 \exp\left(-0.2 \sqrt{\frac{1}{n} \sum_{i=1}^{n} x_i^2}\right) - \exp\left(\frac{1}{n} \sum_{i=1}^{n} \cos(2 \pi x_i)\right) + 20 + e
$$

Where:
- $x_i$ are the coordinates of the input vector \(x\),
- $n$ is the dimensionality of the problem,
- The function combines exponential decay and cosine terms to create the rugged landscape.

### Michalewicz function

The **Michalewicz function** is a popular benchmark function used to evaluate optimization algorithms, particularly in high-dimensional spaces. It is characterized by a complex, highly multimodal landscape filled with many local minima, making it difficult to find the global optimum. The global minimum is typically not at the origin (0,0,...), adding to the challenge for optimization techniques.

The Michalewicz function is defined as:

$$
f(x) = - \sum_{i=1}^{n} \sin(x_i) \cdot \sin^2\left(\frac{i x_i^2}{\pi}\right)
$$

Where:
- $x_i$ is the $i$-th component of the input vector $x$,
- $n$ is the dimensionality of the problem (number of variables),
- The function involves sinusoidal components, making the function oscillate in a complex pattern.


**Global Minimum**: The global minimum occurs at specific non-zero values of $x_i$. The value of the global minimum depends on the dimensionality $n$, but it is generally not located at the origin.

For a problem with two dimensions ($n = 2$), the global minimum is at $x = [2.20, 1.57]$, with the function value $f(x_{\text{min}}) = -1.8013$.

For a problem with two dimensions ($n = 10$), the global minumum function value is $f(x_{\text{min}}) = -9.66015$.

### Schwefel function

The **Schwefel function** has a global minimum at a specific point, but it is filled with many local minima, making it challenging for optimization techniques to find the global optimum. 

The Schwefel function is typically represented as:

$$
f(x) = 418.9829n - \sum_{i=1}^{n} x_i \sin(\sqrt{|x_i|})
$$

Where:
- $x_i$ are the components of the input vector \(x\),
- $n$) is the dimensionality of the problem (number of variables),
- The term $418.9829n$ shifts the function so that the global minimum is non-zero, while the summation term generates the rugged, multi-modal behavior.

**Global Minimum**: The global minimum occurs at $x_i=420.9687$ for all $i$, with $f(x_{\text{min}})=0$. This is at the value where all components of the input vector are equal.

### Schaffer function

The **Shaffer function** is another challenging benchmark function frequently used to test optimization algorithms, particularly for multidimensional optimization problems. It is known for its intricate, multimodal nature, featuring numerous local minima that obscure the path to the global optimum. Like many complex functions, its structure poses significant difficulty for optimization methods, as navigating through the vast number of local optima can be a daunting task.

The Shaffer function is defined as:

$$
f(x) = \sum_{i=1}^{n} \left[ \sin^2\left(\sqrt{x_i^2 + x_{i+1}^2}\right) - 0.5 \right]^2
$$

Where:
- $x_i$ is the $i$-th component of the input vector $x$,
- $n$ is the dimensionality of the problem, typically an even number to pair the variables ($x_i$ with $x_{i+1}$),
- The function involves a sine term applied to the Euclidean distance between consecutive points, which results in a highly oscillatory behavior.

**Global Minimum**: The global minimum occurs at specific non-zero values of $x_i$. Unlike simpler functions, the global optimum is often not at the origin, and the value varies depending on the dimensionality of the problem.

For a problem with two dimensions ($n = 2$), the global minimum is at $x = [0, 0]$, with the function value $f(x_{\text{min}}) = 0$.

For a problem with higher dimensions ($n = 10$), the global minimum continues to have a value of $f(x_{\text{min}}) = 0$, but the location of the minimum changes as the dimensionality increases.

## Conclusions

This study aimed to improve Particle Swarm Optimization (PSO) by integrating key operations from Genetic Algorithms (GA), creating hybrid models designed to balance exploration and exploitation more effectively. Four hybridization techniques were explored: Sequential GAPSO (SGAPSO), Alternating GAPSO (AGAPSO), Parallel GAPSO (PGAPSO), and Island GAPSO (IGAPSO), each with a distinct approach to combining the two algorithms.

Despite the hybridization, the results of the study did not demonstrate consistent improvements across all methods. However, several important considerations have emerged from the analysis:

1. **Dependence on Parameters**: The performance of the hybrid algorithms is highly dependent on the parameters used. It is likely that adjusting key parameters such as mutation rates, crossover probabilities, or the balance between PSO and GA operations could yield better results. Fine-tuning these parameters in the hybrid models might help them perform more effectively in different problem scenarios.

2. **IGAPSO Performance**: Among the hybrid methods tested, **Island GAPSO (IGAPSO)** generally showed more promising results than pure PSO. IGAPSO demonstrated an ability to explore a broader search space and maintain diversity in solutions, which helped avoid premature convergence. This suggests that the island-based model, with its periodic migration of solutions, can provide a more robust optimization process compared to standard PSO, particularly in complex or multimodal optimization problems.

3. **GA Operations' Influence on PSO**: Another crucial finding is the need for further exploration of how GA operations, such as crossover and mutation, interfere with the underlying dynamics of PSO. The interaction between these two algorithms can sometimes disrupt the convergence behavior of PSO, leading to slower optimization or suboptimal solutions. Understanding the precise way in which GA operations affect PSO’s ability to exploit good solutions is essential for refining the hybridization strategy and making it more effective.

Overall, while hybridization demonstrated potential, the results highlight the importance of carefully calibrating the parameters to achieve optimal performance. Future research should focus on better understanding the interplay between GA and PSO, as well as experimenting with different hybridization strategies that could dynamically adjust based on the problem at hand. Further refinement of the GA operations, alongside a more detailed analysis of their effects on PSO’s exploration and exploitation phases, is necessary to fully unlock the potential of hybrid PSO-GA approaches.

In conclusion, while the hybrid PSO-GA algorithms explored in this study did not outperform PSO in all cases, **IGAPSO** stands out as a promising approach, and further investigation into parameter tuning and the interaction between GA operations and PSO is needed to enhance the performance of these hybrid models.
